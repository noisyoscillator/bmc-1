# Plotar dados

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
plt.figure(figsize=(8, 4))
plt.plot(time, disp, label='disp')
plt.plot(time, disp2, label='disp2')
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots (nrows=1, ncols=2, sharex=True, figsize=(8, 4), squeeze=True)
axs[0].plot(time, disp, label='disp')
axs[1].plot(time, aacc, label='aacc')
plt.show()
```

```python
dt = np.mean(np.diff(time))

v = np.diff(disp)/dt

a = np.diff(v)/dt

time = time[:-2]
disp = disp[:-2]
v    =    v[:-1]
aacc = aacc[:-2]
```

```python
fig, axs = plt.subplots (nrows=1, ncols=2, sharex=True, figsize=(8, 4), squeeze=True)
axs[0].plot(time, disp, label='disp')
axs[0].legend()
axs[1].plot(time, aacc, label='aacc')
axs[1].plot(time, a, label='a')
axs[1].legend()
plt.show()
```
